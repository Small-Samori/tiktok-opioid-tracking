```python
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("./test/gpt_annotations_1750639377.240441.txt", sep="\t")
df
```

```python
str_ann = [i[1:-1] for i in list(df["annotation"])]

pred_first = [i[0] for i in str_ann]
pred_second = [i[1] for i in str_ann]
pred_third = [i[2] for i in str_ann]
```

```python
with open("../persons_label-Copy1.txt") as f:
    gt = f.read()
    
gt = gt.split("\n")
print(gt)
```

```python
gt_first = [i[0] for i in gt]
gt_second = [i[1] for i in gt]
gt_third = [i[2] for i in gt]
```

```python
def get_pr(gt, pred):
    tn, fp, fn, tp = confusion_matrix(gt, pred).ravel()
    precision = tp / (tp+fp)
    recall = tp / (tp + fn)
    return precision, recall
```

```python
get_pr(gt_first, pred_first)
```

```python
get_pr(gt_second, pred_second)
```

```python
get_pr(gt_third, pred_third)
```

```python
get_pr(gt_first+gt_second+gt_third, pred_first+pred_second+pred_third)
```
